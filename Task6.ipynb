```python
from task1 import jsim
from task2 import minhash
from task3 import cossim
from task5 import sigsim
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
s1 = [1.0, 2.0, 3.0]
s2 = [2.0, 3.0, 4.0]
```

```python
js = jsim(s1, s2)
cs = cossim(s1, s2)
```

```python
yj = [0]*100
yc = [0]*100
for k in range(1, 101):
    yj[k-1] = abs(sigsim(s1, s2, k) - js)
    yc[k-1] = abs(sigsim(s1, s2, k) - cs)
    
```

```python
x = range(1, 101)
p1 = pyplot.plot(x, yj, '-', label="Jaccard")
p2 = pyplot.plot(x, yc, '-', label="Cosine")
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
pyplot.title( 'Difference to the \'true\' similarity')
pyplot.xlabel( 'Difference' )
pyplot.ylabel( 'k' )
pyplot.show()
```
