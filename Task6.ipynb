```python
from task1 import jsim
from task2 import minhash
from task5 import sigsim
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
s1 = ['a', 'b', 'c']
s2 = ['b', 'c', 'e']
```

```python
js = jsim(s1, s2)
```

```python
y = [0]*100
for k in range(1, 101):
    y[k-1] = abs(sigsim(s1, s2, k) - js)
    
```

```python
x = range(1, 101)
pyplot.plot(x, y, '-' )
pyplot.title( 'Difference to the \'true\' Jaccard similarity')
pyplot.xlabel( 'Difference' )
pyplot.ylabel( 'k' )
pyplot.show()
```
